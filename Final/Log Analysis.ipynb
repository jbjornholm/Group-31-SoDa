```python
# Load the LOG
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import seaborn as sns
```

```python

path2data='C:/Users/frede/OneDrive/Documents/GitHub/Group/Group-31-SoDa/Frederik/Exam/Super Log'
log=pd.read_csv(path2data,';')
log['len']=len(log['redirect_url'])
log['dt'] = pd.to_datetime(log.t,unit='s')
log_movie=log[16:]
log_name=log[:16]
log.head()
```

```python
log_name
```

## MOVIES

```python
plt.figure(figsize=(15,10))
matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25)
sns.lineplot(log_movie.dt,log_movie.delta_t, color='Green')
sns.despine(left=False, right=True, bottom=False)
plt.ylabel('Delta t', fontsize=25)
plt.xlabel('Scraping progress', fontsize=25)

import datetime
plt.savefig("speed, time")
```

```python
# Plot the response size over time.
plt.figure(figsize=(15,10))
matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25)

sns.lineplot(log_movie.dt,log_movie.response_size, color='Green')
sns.despine(left=False, right=True, bottom=False)
plt.ylabel('Size of response', fontsize=25)
plt.xlabel('Scraping progress', fontsize=25)
plt.savefig("size,time")
```

```python
plt.figure(figsize=(15,10))
matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25)
plt.scatter(log_movie.delta_t,log_movie.response_size, color='Green')
sns.despine(left=False, right=True, bottom=False)
plt.ylabel('Size of response', fontsize=25)
plt.xlabel('Delta t', fontsize=25)
plt.savefig("scatter")
```
