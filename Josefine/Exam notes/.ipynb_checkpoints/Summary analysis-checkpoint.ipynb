```python
tokenizer = RegexpTokenizer(r'\w+')
def preprocessing(string):
    lower=string.lower()
    token=tokenizer.tokenize(lower)
    return token

from nltk.probability import FreqDist
```

```python
import string
text_m=df1[df1.Female != 1].reset_index()
stop=stopwords.words('english')
stop1=stop+list(','+'a'+'.'+'"')
def remove_stop(x):
    removed_stop=[]
    for y in x:
        removed_stop.append(' '.join([word for word in y.split() if word not in stop1]))
    return removed_stop
no_stop_m=remove_stop(text_m['Summary'])

su_m=str()
for i in range(6184):
    su_m=su_m + str(no_stop_m[i])

su_m_1=preprocessing(su_m)

fdist = FreqDist(su_m_1)
print(fdist)
male_words = fdist.most_common(33)[3:]
len(male_words)
```

```python
text_f=df1[df1.Female == 1].reset_index()
def remove_stop(x):
    removed_stop=[]
    for y in x:
        removed_stop.append(' '.join([word for word in y.split() if word not in stop]))
    return removed_stop
no_stop_f=remove_stop(text_f['Summary'])

su_f=str()
for i in range(1000):
    su_f=su_f + str(no_stop_f[i])

su_f_1=preprocessing(su_f)

fdist = FreqDist(su_f_1)
print(fdist)
female_words = fdist.most_common(32)[2:]
len(female_words)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)
def grey_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl(0,0%%, %d%%)" % np.random.randint(49,51))


def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=50,
        min_font_size=5, 
        scale=5,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))
    wordcloud.recolor(color_func = grey_color_func)

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(female_words)
show_wordcloud(male_words)
```
