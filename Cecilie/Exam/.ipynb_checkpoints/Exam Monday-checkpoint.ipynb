```python
import pandas as pd

df = pd.read_csv("/Users/cecil/OneDrive/Dokumenter/sds2019_2/Group-31-SoDa/Data/movie_1.csv", index_col=0)
   
df2 = df.copy()    
df2.head(1)
```

```python
#Cleaning genre
df2['Genre 1'] = df["Genre 1"].str.replace(" ", "")
df2["Genre 2"] = df["Genre 2"].str.replace(" ", "")
df2["Genre 3"] = df["Genre 3"].str.replace(" ", "")
df2

```

```python
test_df = df2[['Female', 'Genre 1', 'Genre 2', 'Genre 3']]
test_df.head(1)
```

```python
genre_1 = test_df[['Female', 'Genre 1']]
genre_2 = test_df[['Female', 'Genre 2']]
genre_3 = test_df[['Female', 'Genre 3']]

genre_1x = genre_1.rename(columns = {'Genre 1': 'Genre'})
genre_2x = genre_2.rename(columns = {'Genre 2': 'Genre'})
genre_3x = genre_3.rename(columns = {'Genre 3': 'Genre'})


all_genres = pd.concat([genre_1x, genre_2x, genre_3x])  
all_genres
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
#all_genres.plot()

all_genres["Genre"].value_counts().T.plot(kind='bar')
plt.show()
```

```python
# Plot for all the genres
plt.figure(figsize=(20,5))
sns.set(style='whitegrid')
f1 = sns.countplot(x="Genre", data=all_genres, hue="Female", order = all_genres['Genre'].value_counts().index, hue_order = all_genres['Female'].value_counts().index)
plt.xlabel("Genre", fontsize=13, color="Black")
plt.ylabel("Count", fontsize=13, color="Black")
plt.title("Genre Distribution by Gender", fontsize=14, color="Black", fontweight="bold")
plt.legend(loc='upper right', frameon=False)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.set(style='whitegrid')
f1 = sns.countplot(x="Genre", data=all_genres, hue="Female", order = all_genres['Genre'].value_counts()[:10].index, hue_order = all_genres['Female'].value_counts().index)
plt.xlabel("Genre", fontsize=13, color="Black")
plt.ylabel("Count", fontsize=13, color="Black")
plt.title("Genre Distribution by Gender - Top 10 (Absolutte numbers)", fontsize=14, color="Black", fontweight="bold")
plt.legend(loc='upper right', frameon=False)
plt.show()

#Ad total amount of males vs females in the graph below in a pie chart
```
