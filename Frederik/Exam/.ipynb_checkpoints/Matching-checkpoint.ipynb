```python
rev=str(done['Top Star'])
tokenizer = nltk.tokenize.TweetTokenizer()

def preprocessing(string):
    token=tokenizer.tokenize(string)
    return token
def count_dictionary(doc,dic):
    doc_new=[ x for x in doc if x in dic]
    return len(doc_new)

check=pd.DataFrame([range(10000)])
done['Male Counter']=done['Top Star'].apply(preprocessing).apply(count_dictionary, dic=male)
done['Female Counter']=done['Top Star'].apply(preprocessing).apply(count_dictionary, dic=female)

done
```
