## Pie Chart

```python
# Gender of Actors and Ratio calculation

print(df["Female"].value_counts()) # 0 = Male, 1 = Female
print(len(df["Female"]))
ratios = df["Female"].value_counts("0")
ratios
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Male', 'Female'
#sizes = [15, 30] # ratios female and male
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Female')

fig1, ax1 = plt.subplots()
wedges, texts, autotexts = ax1.pie(ratios, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=180, colors = ["green",'darkseagreen'], textprops=dict(color="white"), wedgeprops   = { 'linewidth' : 1,'edgecolor' : "white" })
plt.setp(autotexts, size=15, weight="bold")
plt.legend(wedges, labels, loc="best", frameon=False)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

## Genre

```python
test_df = df2[['Female', 'Genre 1', 'Genre 2', 'Genre 3']]
test_df.head(1)
```

```python
genre_1 = test_df[['Female', 'Genre 1']]
genre_2 = test_df[['Female', 'Genre 2']]
genre_3 = test_df[['Female', 'Genre 3']]

genre_1x = genre_1.rename(columns = {'Genre 1': 'Genre'})
genre_2x = genre_2.rename(columns = {'Genre 2': 'Genre'})
genre_3x = genre_3.rename(columns = {'Genre 3': 'Genre'})


all_genres = pd.concat([genre_1x, genre_2x, genre_3x])  
all_genres.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
#all_genres.plot()

all_genres["Genre"].value_counts().T.plot(kind='bar', color="green")
plt.show()
```

### Plot Genre - Total

```python
# Plot for all the genres
plt.figure(figsize=(20,5))
sns.set(style='whitegrid')
f1 = sns.countplot(x="Genre", data=all_genres, hue="Female", order = all_genres['Genre'].value_counts().index, hue_order = all_genres['Female'].value_counts().index, palette="Greens")
plt.xlabel("Genre", fontsize=13, color="Black")
plt.ylabel("Count", fontsize=13, color="Black")
plt.title("Genre Distribution by Gender", fontsize=14, color="Black", fontweight="bold")
plt.legend(loc='upper right', frameon=False)
plt.show()
```

### Plot Genre Top 10

```python
plt.figure(figsize=(20,5))
sns.set(style='whitegrid')
f1 = sns.countplot(x="Genre", data=all_genres, hue="Female", order = all_genres['Genre'].value_counts()[:10].index, hue_order = all_genres['Female'].value_counts().index, palette="Greens")
plt.xlabel("Genre", fontsize=13, color="Black")
plt.ylabel("Count", fontsize=13, color="Black")
#plt.title("Genre Distribution by Gender - Top 10 (Absolutte numbers)", fontsize=14, color="Black", fontweight="bold")
plt.legend(loc='upper right', frameon=False)
plt.show()

#Ad total amount of males vs females in the graph below in a pie chart
```
